## Test Driven Development

**Motivation**: Why should I write the tests first?
- Early Bug Detection: Bugs are caught immediately when introduced, not later in development
- Tighter Feedback Loop: Mistakes are found and fixed almost instantly
- Prevention Over Cure: Prevents bugs from accumulating and becoming harder to fix
- Faster Development: Less time spent debugging complex issues later in the process
- Better Code Quality: Continuous validation ensures code works correctly from the start
- Reduced Debugging Time: Eliminates long gaps between creating and discovering defects

## Advantages of TDD

- Improved software quality (fewer bugs)
- Reduced debugging time
- Leads to clean and modular design
- Allows for safe code changes
- Executable documentation
- Keeps track of progress
- Predictable development time

___

## TDD Cycle

1. Write a test
2. Run the tests and watch it fail
3. Write the code to make it pass
4. Run the tests and see it works
5. Refactor code

___

## Getting Started

#### Which tests should I write?

- Writing a test plan
    - What behavior should be implemented?
    - What output do we expect?
    - Input constraints
    - Failure conditions
- Incremental Design
    - Unit by unit
    - Non regression
    - Test the behaviors not the methods !

## Fundations

#### Unit Test
Verify the behavior of a code unit.
1. (Optional) Set up the execution context (Given)
2. Some statements to invoke the behavior you want to check (When)
3. Some statements to check the expected outcome (Then)

They can be grouped in a test suite.

#### Using tests to describe behavior
Given a meaningful name + test statement test statement

## Example 

```cpp
// Let DocTest provide main():
#define DOCTEST_CONFIG_IMPLEMENT_WITH_MAIN
#include "doctest.h"
#include "passgrid.hh"
TEST_SUITE("grid") {
    TEST_CASE("1: Size of the grid are initialized by construction") {
        PassGrid p0(0, 0);
        CHECK(p0.getW() == 0);
        CHECK(p0.getH() == 0);
        PassGrid p1(10, 8);
        CHECK(p1.getW() == 10);
        CHECK(p1.getH() == 8);
    }
    TEST_CASE("2: copy same grid give two identical grids") {
        PassGrid p0(0, 0);
        PassGrid p1(p0);
        CHECK(!notEqualGrids(p0, p1));
    }
}
```
